```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt

# Create plots for data evaluation
```

```python
%matplotlib inline
data_model = pd.DataFrame()
os.chdir("~/SS20-Optimizing3DVideoMaterial/")
directories = {'data'}
for i in directories:
    files = [i for i in glob.glob(f'{i}/*.csv')]
    #files = {'Latency_1.csv'}
    for file in files:
        print(file)
        df = pd.read_csv(file, sep=';')
        df['file'] = file
        data_model = pd.concat([data_model, df['Latency']])
```

```python
print(data_model.quantile(.1))
print(data_model.quantile(.9))
print(data_model.quantile(.5))

plt.subplots(figsize=(20,2))
plt.title('Latency', fontsize=15)
plt.xlabel('Time in ms', fontsize=15)
data_model.boxplot(vert=False,showfliers=False)


plt.subplots(figsize=(20,2))
plt.title('Latency', fontsize=15)
plt.xlabel('Time in ms', fontsize=15)
data_model.boxplot(vert=False,showfliers=True)


data_model.plot()
plt.title('Graph')
plt.subplots(figsize=(20,2))
```
